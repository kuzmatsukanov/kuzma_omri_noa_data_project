```python
# Shows how many files in the images folder
def list_files(dir):
    """
    Returns list of filepaths in the given directory (dir) excluding hidden files
    """
    file_list = []
    for root, dirs, files in os.walk(dir):
        for file in files:
            if not file.startswith('.'):
                file_list.append(os.path.join(root, file))
    return file_list
file_lst = list_files('drive/MyDrive/ds_project_dataset/images')
len(file_lst)
```

## Image encoder

```python
def get_img_embedding(img_path, encoder, TARGET_SIZE=(80, 80)):
    """
    Returns embedding vector of the image
    img_path: filepath, e.g., "samples/sample_dress.jpg"
    encoder: model for embedding, eg, ResNet50
    TARGET_SIZE: resize to the target size, e.g. (80, 80)
    returns embedded_input_img (numpy.ndarray): embedded vector of the image
    """ 
    # Load the image
    image = load_img(path=img_path, target_size=TARGET_SIZE)
    
    # Convert the image to numpy arrays
    input_img = img_to_array(image)
    
    # Turning the image into embedding vector
    embedded_input_img = encoder.predict(input_img[np.newaxis, :]).flatten()
    
    # Reduce the size of memory usage
    embedded_input_img = embedded_input_img.astype('float16')
    return embedded_input_img
```

```python
# Define file path and column headers
file_path = 'table_embedded_vectors.csv'
headers = ['id', 'vector']

# Create a file ID | embedded vector
with open(file_path, 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(headers)
```

```python
# Define target size of the image
TARGET_SIZE = (80, 80)

# Define the model for embedding using the model ResNet50
encoder = ResNet50(weights="imagenet", include_top=False)
```

```python
for img_path in tqdm(file_lst):
    # Get embedding vector of the image
    embedded_input_img = get_img_embedding(img_path=img_path, encoder=encoder, TARGET_SIZE=TARGET_SIZE)
    row_to_write = (img_path[-13:-4], list(embedded_input_img))
    
    # Append embedded vector with correspondent ID to the file
    with open(file_path, 'a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(row_to_write)
```

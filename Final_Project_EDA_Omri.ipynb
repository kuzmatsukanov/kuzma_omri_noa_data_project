```python
group_names = ['Garment Upper body', 'Garment Lower body', 'Garment Full body', 'Accessories', 'Underwear', 'Shoes', 'Swimwear', 'Socks & Tights', 'Nightwear']
```

```python
filtered_df = df.iloc[df.index[df['product_group_name'].isin(group_names)].tolist()]
```

```python
filtered_df.columns
```

```python
colour_group_name = filtered_df['colour_group_name']
```

```python
grouped = colour_group_name.value_counts()[:20]
```

```python
colors = sns.color_palette('pastel')
plt.figure(figsize=(15,15))
plt.pie(grouped.values, labels = grouped.index, colors = colors, autopct='%.0f%%')
plt.title("Color Group Names Distribution")
plt.tight_layout()
plt.show()
```

```python
perceived_colour_master_name = filtered_df['perceived_colour_master_name']
```

```python
grouped2 = perceived_colour_master_name.value_counts()[:16]
```

```python
plt.figure(figsize=(15,15))
plt.pie(grouped2.values, labels = grouped2.index, colors = colors, autopct='%.0f%%')
plt.title("Perceived Color Master Names Distribution")
plt.tight_layout()
plt.show()
```

```python
filtered_df2 = filtered_df[(filtered_df.perceived_colour_master_name != 'Yellowish Green') & (filtered_df.perceived_colour_master_name != 'Bluish Green')]
```

```python
plt.figure(figsize=(20,15))
plt.scatter('perceived_colour_master_name', 'colour_group_name', data=filtered_df2, alpha = 0.3)
plt.title(f'perceived_colour_master_name vs colour_group_name')
plt.xlabel('perceived_colour_master_name')
plt.ylabel('colour_group_name')
plt.tight_layout()
plt.show();
```

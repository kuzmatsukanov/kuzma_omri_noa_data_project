```python
plt.figure(figsize=(10,5))
df['product_group_name'].value_counts().sort_index().plot(kind = 'barh')
plt.xticks(rotation=0)
plt.title(f'target Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,5))
df['product_group_name'].value_counts()[:9].sort_index().plot(kind = 'barh')
plt.xticks(rotation=0)
plt.title(f'Clothing Items Classes Distribution')
plt.show()
```

```python
plt.figure(figsize=(10,50))
df['product_type_name'].value_counts().sort_index().plot(kind = 'barh')
plt.xticks(rotation=0)
plt.title(f'Clothing Item Types Classes Distribution')
plt.show()
```
